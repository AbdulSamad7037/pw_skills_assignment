2.Explain the difference between a for loop and a while loop in Python.


Structure:

For Loop: It is typically used when you know in advance how many times you want to execute a block of code. It iterates over a sequence (e.g., a range of numbers) or any iterable object.

While Loop: It is used when the number of iterations is not known in advance and depends on a condition. The loop continues executing as long as the specified condition is True.


Initialization:

For Loop: The loop variable is initialized within the loop header. The loop iterates over a predefined sequence or iterable, and the variable takes on the values from that sequence.

While Loop: The loop variable or condition is typically initialized before the loop. The loop continues as long as the specified condition is True.

Termination:

For Loop: The loop terminates when it has iterated over the entire sequence or iterable.

While Loop: The loop terminates when the specified condition becomes False.

Utilize a for loop to generate the remaining numbers in the sequence. The loop iterates series_length times.

Print Next Number:

Print the calculated next number. This represents the next term in the Fibonacci sequence.
Update Values:

Completion:

Allow the loop to run its course, generating the specified length of the Fibonacci sequence.

13. Write a Python program to reverse a list using a for loop.

```python
def reverse_list(input_list):
    reversed_list = []

    # Iterate over the elements of the input list in reverse order
    for i in range(len(input_list) - 1, -1, -1):
        reversed_list.append(input_list[i])

    return reversed_list

# Example usage:
original_list = [1, 2, 3, 4, 5]
reversed_result = reverse_list(original_list)

print(f"Original List: {original_list}")
print(f"Reversed List: {reversed_result}")


```

reverse_list is a function that takes an input list and returns a new list containing the elements of the input list in reverse order.

The for loop iterates over the indices of the input list in reverse order using range(len(input_list) - 1, -1, -1). Inside the loop, each element is appended to the reversed_list.



14. Write a Python program to find the common elements between two lists using a for loop.

```python
def find_common_elements(list1, list2):
    common_elements = []

    # Iterate over elements in the first list
    for element in list1:
        # Check if the element is also in the second list
        if element in list2:
            common_elements.append(element)

    return common_elements


list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]

common_elements_result = find_common_elements(list1, list2)

print(f"List 1: {list1}")
print(f"List 2: {list2}")
print(f"Common Elements: {common_elements_result}")
```

def find_common_elements(list1, list2):: This line defines a function named find_common_elements that takes two lists (list1 and list2) as parameters.

common_elements = []: This line initializes an empty list named common_elements. This list will be used to store the common elements between the two input lists.

for element in list1:: This line initiates a for loop that iterates over each element in the first list (list1).

Inside the loop:

if element in list2:: This line checks if the current element from the first list is also present in the second list (list2).
common_elements.append(element): If the element is common (i.e., it is found in both lists), it is appended to the common_elements list.
return common_elements: After the loop, the function returns the list of common elements.

18. Write a Python program to remove duplicates from a list using a for loop.

```python
# Python code to remove duplicate elements
def Remove(duplicate):
	final_list = []
	for num in duplicate:
		if num not in final_list:
			final_list.append(num)
	return final_list
	

duplicate = [2, 4, 10, 20, 5, 2, 20, 4]
print(Remove(duplicate))
```

The function Remove takes a list called duplicate as an argument.

Inside the function, there is an empty list named final_list initialized. This list will store the unique elements from the original list.

The function then uses a for loop to iterate through each element (num) in the input list (duplicate).

Inside the loop, there is an if statement that checks whether the current element (num) is not already present in the final_list. If it's not in the list, it means the element is unique, and it is appended to the final_list.

The function continues to iterate through all elements in the input list, adding only the unique elements to final_list.

Finally, the function returns the final_list, which contains only the unique elements from the original list.

21. Write a Python program that checks if a given year is a leap year using a for loop.


```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0):
        return True
    else:
        return False

# Get the number of years to check
num_years = int(input("Enter the number of years to check: "))

years = []
# Get user input for each year and store in the list
for i in range(num_years):
    year = int(input(f"Enter year {i + 1}: "))
    years.append(year)

# Check if each year in the list is a leap year using a for loop
for year in years:
    if is_leap_year(year):
        print(f"{year} is a Leap year")
    else:
        print(f"{year} is not a Leap year")

```

## 1.Explain the concept of recursion in Python. How does it differ from iteration?


Recursion is a programming technique where a function calls itself in order to solve a problem. It's a powerful concept in computer science and often used in algorithms and data structures. In Python, recursion is implemented by defining a function that calls itself within its own definition.

Recursion differs from iteration (looping) in how it approaches solving a problem. While iteration involves executing a set of instructions repeatedly, recursion involves breaking down a problem into smaller subproblems and solving each subproblem recursively until reaching a base case.

## 2.Write a Python program to calculate the factorial of a number using recursion.


```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# Test the function
num = int(input("Enter a non-negative integer: "))
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    print("Factorial of", num, "is", factorial(num))
```

## 3.Create a recursive Python function to find the nth Fibonacci number.


```python
def fibonacci(n):
    if n <= 0:
        return "Invalid input. Fibonacci sequence starts from 1."
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Test the function
n = int(input("Enter the value of n to find the nth Fibonacci number: "))
print("The", n, "th Fibonacci number is:", fibonacci(n))
```

## 10. Write a recursive Python function to check if a string is a palindrome.


```python
def is_palindrome(s):
    # Base case: if the length of the string is 0 or 1, it's a palindrome
    if len(s) <= 1:
        return True
    # Recursive case: check if the first and last characters are the same,
    # and then recursively check the substring without the first and last characters
    else:
        return s[0] == s[-1] and is_palindrome(s[1:-1])

# Test the function
input_string = input("Enter a string: ")
if is_palindrome(input_string):
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

## 11.Create a recursive Python function to generate all possible combinations of a list of elements.


```python
def generate_combinations(elements, current=[], index=0, combinations=[]):
    if index == len(elements):
        combinations.append(current[:])  # Append a copy of the current combination
    else:
        # Include the current element in the combination
        current.append(elements[index])
        generate_combinations(elements, current, index + 1, combinations)
        
        # Exclude the current element from the combination
        current.pop()
        generate_combinations(elements, current, index + 1, combinations)

    return combinations

# Test the function
elements = [1, 2, 3]
all_combinations = generate_combinations(elements)
print("All possible combinations:", all_combinations)
```

```python
def factorial(n):
    """
    Calculate the factorial of a number.

    This function returns the factorial of a non-negative integer n.
    The factorial of a number is the product of all positive integers
    less than or equal to the number.

    Parameters:
    n (int): A non-negative integer whose factorial is to be calculated.

    Returns:
    int: The factorial of the input number.
    """
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
```
